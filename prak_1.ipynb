```python
import matplotlib.pyplot as plt

def lis(arr):
    n = len(arr)
    # Inisialisasi array LIS, setiap elemen minimal membentuk subsequence panjang 1
    lis = [1] * n

    # Menjelaskan proses perhitungan LIS
    print("Tahap Perhitungan LIS:")
    for i in range(1, n):
        for j in range(0, i):
            if arr[i] > arr[j] and lis[i] < lis[j] + 1:
                lis[i] = lis[j] + 1

        # Cetak status array lis setelah setiap iterasi
        print(f"Setelah memproses arr[{i}] = {arr[i]}: {lis}")

    # Mencari panjang maksimum dari LIS
    max_length = max(lis)
    print(f"\nPanjang maksimum LIS: {max_length}")

    # Membangun subsequence LIS dari belakang
    result = []
    current_length = max_length
    for i in range(n - 1, -1, -1):
        if lis[i] == current_length:
            result.append(arr[i])
            current_length -= 1

    # Mengembalikan hasil LIS yang telah dibalik agar urutannya benar
    return result[::-1]

def visualize(arr, lis_result):
    # Buat plot untuk visualisasi
    plt.figure(figsize=(10, 6))

    # Plot array asli
    plt.plot(range(len(arr)), arr, marker='o', linestyle='-', label='Array Asli', color='blue')

    # Tandai elemen LIS
    lis_indices = [arr.index(num) for num in lis_result]
    lis_values = lis_result
    plt.plot(lis_indices, lis_values, marker='o', linestyle='-', color='red', label='LIS', linewidth=2)

    # Tambahkan label dan legend
    plt.title("Visualisasi Largest Monotonically Increasing Subsequence")
    plt.xlabel("Index")
    plt.ylabel("Nilai")
    plt.legend()
    plt.grid(True)

    # Tampilkan plot
    plt.show()

# Input deret bilangan
arr = [4, 1, 13, 7, 0, 2, 8, 11, 3]
print(f"Array Input: {arr}\n")

# Proses LIS
result = lis(arr)
print(f"\nLargest Monotonically Increasing Subsequence: {result}")

# Visualisasi hasil
visualize(arr, result)
```
